```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Simulated blood sugar data (mg/dL)
np.random.seed(42)
normal_group = np.random.normal(95, 10, 120)   # Normal range
diabetic_group = np.random.normal(150, 15, 80) # Diabetic range
blood_sugar = np.concatenate([normal_group, diabetic_group])

# KDE Plot
sns.kdeplot(blood_sugar, shade=True, color='teal')
plt.title("Blood Sugar Level Distribution (KDE Plot)")
plt.xlabel("Blood Sugar Level (mg/dL)")
plt.ylabel("Density")
plt.show()
```
